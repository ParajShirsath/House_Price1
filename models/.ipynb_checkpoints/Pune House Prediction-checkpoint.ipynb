# Feature Engineering

```python
### Size
```

```python
df[['size']].boxplot()
```

```python
q1=df['size'].quantile(0.25)
q2=df['size'].quantile(0.50)
q3=df['size'].quantile(0.75)


iqr = q3 -q1

upper_tail = q3+ 1.5*iqr
lower_tail = q1- 1.5*iqr

print("q1>>", q1)
print("q2>>", q2)
print("q3>>", q3)
print("upper_tail>>", upper_tail)
print("lower_tail>>", lower_tail)


size_df = df[['size']].loc[(df['size'] < upper_tail) & (df['size'] > lower_tail)].median()[0]

size_df

df.loc[(df['size'] > upper_tail) | (df['size'] < lower_tail), 'size'] = size_df

df[['size']].boxplot()
```

```python
### Bath
```

```python
df[['bath']].boxplot()
```

```python
q1=df['bath'].quantile(0.25)
q2=df['bath'].quantile(0.50)
q3=df['bath'].quantile(0.75)


iqr = q3 -q1

upper_tail = q3+ 1.5*iqr
lower_tail = q1- 1.5*iqr

print("q1>>", q1)
print("q2>>", q2)
print("q3>>", q3)
print("upper_tail>>", upper_tail)
print("lower_tail>>", lower_tail)


df['bath'].mean()

df['bath'].median()

size_df = df[['bath']].loc[(df['bath'] < upper_tail) & (df['bath'] > lower_tail)].median()[0]

size_df

df.loc[(df['bath'] > upper_tail) | (df['bath'] < lower_tail), 'bath'] = size_df

df[['bath']].boxplot()
```

## Model Training

```python
x=df.drop('price',axis=1)
x
```

```python
y=df['price']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
## Ceating Model Instance
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

## Metric Evaluation

```python
y_pred_test = model.predict(x_test)
y_pred_test
```

```python
r2=r2_score(y_test,y_pred_test)
r2
```

```python
residual=y_test-y_pred_test
residual
```

```python
# Testing Dataset
y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred_test)
print("MSE :",mse)

rmse = np.sqrt(mse)
print("RMSE :",rmse)

mae = mean_absolute_error(y_test, y_pred_test)
print("MAE :",mae)

r2 = r2_score(y_test, y_pred_test)
print("R Squared :",r2)
```

```python
# training Data Set
y_pred_train = model.predict(x_train)

mse = mean_squared_error(y_train, y_pred_train)
print("MSE :",mse)

rmse = np.sqrt(mse)
print("RMSE :",rmse)

mae = mean_absolute_error(y_train, y_pred_train)
print("MAE :",mae)

r2 = r2_score(y_train, y_pred_train)
print("R Squared :",r2)
```
